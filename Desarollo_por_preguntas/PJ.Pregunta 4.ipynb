# 4 ¿Cuál es el producto en el que somos más caros ?, ¿y el más barato? ¿Y DIA?¿Y Carrefour?

* Antes de responder a la pregunta hay que objetivizar el concepto "mas caro", ya que podrría ser el producto mas caro o mas barato con respecto a nuestra competencia o el preoducto mas caro y mas barato que vendemos nosotros y nuestros competidores

* Como se venden distintos productos entre supermercados y muchos de ellos no se pueden comparar con otros, vamos a seleccionar aquellos productos que venden mas caros y mas baratos en cada supermercado.

```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("raw_data.csv")
```

```python
df_mercadona= df[df["id_reference_supermarket"]=="M1"]
df_carrefour= df[df["id_reference_supermarket"]=="M2"]
df_dia= df[df["id_reference_supermarket"]=="M3"]
```

```python
df.info()
```

```python
display(df_mercadona.sort_values(by="reference_price",ascending=False).head(2))
display(df_carrefour.sort_values(by="reference_price",ascending=False).head(2))
display(df_dia.sort_values(by="reference_price",ascending=False).head(2))
```

#### Inciso

* Al observar el precio de referencia, el precio y la unidad de referencia vemos que los datos para estos valores están mal si tenemos en cuenta el precio de referencia como indicador del producto mas caro, por ello vamos a ver cual es el producto mas caro por precio y haremos unos gráficos de dispersión para ver si tienen sentido. 

```python
display(df_mercadona.sort_values(by="price",ascending=False).head(2))
display(df_carrefour.sort_values(by="price",ascending=False).head(2))
display(df_dia.sort_values(by="price",ascending=False).head(2))
```

#### Inciso

* Los precios son mas coherentes que los precios de referencia, tal y como observamos en los gráficos de dispersión. El problema que tenemos en el caso del dataset del mercadona es que algunos productos están vendidos al granel y esto hace que el precio de su conjunto sea muy caro y no sea representativo. Es muy raro que una persona compre en el Mercadona 100kg de carabineros. Por lo que vamos a disponernos a hacer un filtrado de los productos del mercadona para seleccionar el producto que se ajuste más a nuestros criterios. 

```python
df_mercadona_price_filter = df_mercadona[df_mercadona["description"]!="Granel"]
df_mercadona_price_filter.sort_values(by="price",ascending=False).head(2)
```

#### Inciso

* Una vez que tenemos los productos mas caros y que se ha verificado su coherencia, haremos lo mismo pero con los más baratos

```python
display(df_mercadona.sort_values(by="price").head(2))
display(df_carrefour.sort_values(by="price").head(2))
display(df_dia.sort_values(by="price").head(2))
```

#### Inciso

* Si seguimos el criterio establecido para responder a cual es el producto mas caro, los resultados de arriba serían válidos, pero observamos que dos de los tres resultantes tienen un precio muy bajo porque se venden en unidades muy pequeñas, siendo su precio de referencia bastante superios. En el caso de los productos más bajos de precio la relación entre el precio de referencia, el precio y la unidad está bien, por lo que además de mostrar estos resultados como los más baratos también podemos mostrar aquellos que guarden una relación similar entre precio y precio de referencia.

```python
df_mercadona_price_filter2 = df_mercadona[(df_mercadona["description"]!="Granel") & (df_mercadona["reference_price"]<0.5)]
df_mercadona_price_filter2.sort_values(by="price").head(2)
```

```python
df_dia_price_filter2 = df_mercadona[(df_mercadona["description"]!="Granel") & (df_mercadona["reference_price"]<0.5)]
df_dia_price_filter2.sort_values(by="price").head(2)
```
