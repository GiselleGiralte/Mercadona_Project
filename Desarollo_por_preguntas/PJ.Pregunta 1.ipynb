# 1. ¿Está su API actualizada? ,¿cuántos productos faltan?

### Primero obtenemos los datos de la API de mercadona para crear un CSV con todos los productos y poder compararlo con el raw_data.csv

```python
import requests as rq
import pandas as pd
```

```python
base="https://tienda.mercadona.es/api/categories/"
```

```python
ra= rq.get(base)
```

```python
raj= ra.json()
```

```python
pdra=pd.json_normalize(raj)
pdra
```

```python
pdra2=pd.DataFrame(pdra["results"][0])
pdra2.head()
```

```python
all_categories = pd.DataFrame(
{"id" : [int()], 
"name" : [str()], 
"order" : [int()],
"layout" : [int()],
"published" : [bool()],
"is_extended" : [bool()]}, 
index = [])

```

```python
for i in pdra2["id"]:
    try:
        all_categories=pd.concat([all_categories,pd.DataFrame(pdra2["categories"][i])])
    except:
        continue

all_categories.reset_index()
```

```python
new_row={"id":26,"name":"Limpieza y hogar","order":304,"layout":1,"published":False,"is_extended":False}
all_categories_complete=pd.concat([all_categories,pd.DataFrame(new_row, index=[151])])
all_categories_complete
```

```python
all_categories_complete.isna().sum()
```

```python
all_categories_api2 = pd.DataFrame(
{"id" : [int()], 
"name" : [str()], 
"order" : [int()],
"layout" : [int()],
"published" : [bool()],
"is_extended" : [bool()]}, 
index = [])
```

```python
#for i in all_categories["id"]:
    #all_categories_api2=pd.concat([all_categories_api2,pd.json_normalize(rq.get(f"https://tienda.mercadona.es/api/categories/{i}").json())])
#all_categories_api2    
```

```python
for i in all_categories_complete["id"]:
    try:
        all_categories_api2=pd.concat([all_categories_api2,pd.json_normalize(rq.get(f"https://tienda.mercadona.es/api/categories/{i}").json())])
    except:
        continue
```

```python
all_categories_api2.isna().sum()
```

```python
all_categories_api2.drop(["level_0","index"],axis=1,inplace=True)
```

```python
all_categories_api2.head()
```

```python
all_df_products = pd.DataFrame(
{"id" : [int()], 
"name" : [str()], 
"order" : [int()],
"layout" : [int()],
"products" : [str()],
"published" : [bool()],
"is_extended" : [bool()]}, 
index = [])
```

```python
for i in all_categories_api2["categories"]:
    try:
        all_df_products=pd.concat([all_df_products,pd.DataFrame(i)])
    except:
        continue

all_df_products
```

```python
all_df_products.isna().sum()
```

```python
all_products = pd.DataFrame(
{"id" : [int()], 
"slug" : [str()], 
"limit" : [int()],
"badges" : [int()],
"packaging" : [str()],
"published" : [bool()],
"share_url" : [bool()],
"thumbnail" : [bool()],
"categories" : [bool()],
"display_name" : [bool()],
"price_instructions" : [bool()]}, 
index = [])
```

```python
for i in all_df_products['products']:
    try:
        all_products=pd.concat([all_products,pd.DataFrame(i)])
    except:
        continue

all_products.head()
```

```python
all_products.isna().sum()
```

```python
list_all_products=[i for i in all_products["display_name"]]
pd.DataFrame(list_all_products,columns=["display_name"])
```

```python
all_mercadonaApi_products_price = pd.DataFrame(
{"iva" : [int()], 
"is_new" : [str()], 
"is_pack" : [int()],
"pack_size" : [int()],
"unit_name" : [str()],
"unit_size" : [bool()],
"bulk_price" : [bool()],
"unit_price" : [bool()],
"approx_size" : [bool()],
"size_format" : [bool()],
"total_units" : [str()], 
"unit_selector" : [int()],
"bunch_selector" : [int()],
"drained_weight" : [str()],
"selling_method" : [bool()],
"price_decreased" : [bool()],
"reference_price" : [bool()],
"min_bunch_amount" : [bool()],
"reference_format" : [bool()],
"increment_bunch_amount" : [bool()]}, 
index = [])
```

```python
for i in all_products["price_instructions"]:
    try:
        all_mercadonaApi_products_price=pd.concat([all_mercadonaApi_products_price,pd.DataFrame(i,index=[0])])
    except:
        continue
        
all_mercadonaApi_products_price
```

```python
all_mercadonaApi_products_price.reset_index(inplace=True)
```

```python
final_M1_Api_df=pd.concat([pd.DataFrame(list_all_products,columns=["display_name"]),all_mercadonaApi_products_price],axis=1)
```

```python
final_M1_Api_df.drop(["index"],axis=1).head()
```

```python
final_M1_Api_df.to_csv("final_M1_Api_ProductPrice.csv")
```

### Justificación 

* Para ver si los productos de la API del mercadona están actualizados se ha creado un csv con todos los productos únicos que hay en esta. Se han tenido en cuenta aquellos productos que se venden en formatos distintos como productos únicos ya que se aplica el mismo criterio para el raw_data.csv

* Para contabilizar productos únicos en el raw_data.csv se ha utilizado el product_id, que es único para cada producto ya que muchos se repiten debido a que tienen fechas de inserción distintas pero siguen siendo el mismo producto.

* Para comparar los precios se han tenido en cuenta varios criterios con respecto a los datos que había en el raw_data.csv ya que según la fecha de inserción el precio varía para un mismo producto. Por ello, se ha comparado la media, la mediana, la desviación típica y los cuartiles(0.25-0.75) para todas las fechas y para aquellos en que el precio de referencia era el máximo.

### Respuesta

* La API está desactualizada ya que en el raw_data.csv hay más productos y ademas los precios son mas caros siendo la diferencia aún mas notable cuando escogemos los precios máximos que han tenido los diferentes productos.
